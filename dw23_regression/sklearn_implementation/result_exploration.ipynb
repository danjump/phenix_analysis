```python
import pandas as pd
import numpy as np
import shelve
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
f = shelve.open('results/sk0.18/d30w50_a0_c0.shelf', flag='r')
hist = f['hist']
dfp = f['dfp']
df_chi2p = f['df_chi2p']

print 'hist:', hist.columns
print 'dfp:', dfp.columns
print 'df_chi2p:', df_chi2p.columns
```

```python
wness = hist['wness_bin'].values
dw23 = hist['dw23_bin'].values
entries = hist['entries'].values

p_wness = dfp['wness'].values
p_dw23 = dfp['dw23'].values
p_entries = dfp['entries'].values


plt.figure(num=1, figsize=(9, 7)) 
ax = plt.subplot(projection='3d')
ax.scatter(wness, dw23, entries, color='b')
ax.scatter(p_wness, p_dw23, p_entries, color='r')
ax.view_init(elev=0, azim=-180)
ax.set_xlabel('wness')
ax.set_ylabel('dw23')
ax.set_zlabel('counts')
```
